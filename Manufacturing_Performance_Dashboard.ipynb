```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
# عدد الصفوف المطلوب
num_rows = 500

# تحديد تاريخ البدء
start_date = datetime(2023, 1, 1)

# توليد قائمة بالتواريخ
date_list = [start_date + timedelta(days=x) for x in range(num_rows)]

# توليد البيانات لكل عمود
data = {
    'Date': date_list,
    'Shift': [random.choice(['صباحية', 'مسائية', 'ليلية']) for _ in range(num_rows)],
    'Machine_ID': [f'M_{random.randint(1, 10)}' for _ in range(num_rows)],
    'Temperature': np.random.normal(75, 5, num_rows).round(2),  # متوسط 75°C بانحراف معياري 5
    'Vibration': np.random.normal(0.5, 0.1, num_rows).round(2),  # متوسط 0.5 mm/s بانحراف معياري 0.1
    'Operating_Hours': np.random.uniform(20, 24, num_rows).round(2),  # بين 20 و 24 ساعة
    'Energy_Consumption': np.random.uniform(400, 600, num_rows).round(2),  # بين 400 و 600 kWh
    'Production_Quantity': np.random.randint(900, 1100, num_rows),  # بين 900 و 1100 وحدة
    'Scrap_Rate': np.random.uniform(0.5, 2.0, num_rows).round(2),  # بين 0.5% و 2.0%
    'Defect_Rate': np.random.uniform(0.5, 1.5, num_rows).round(2),  # بين 0.5% و 1.5%
    'Unplanned_Downtime': np.random.randint(0, 3, num_rows),  # بين 0 و 2 مرات
    'Downtime_Duration': np.random.uniform(0, 120, num_rows).round(2),  # بين 0 و 120 دقيقة
    'Maintenance_Cost': np.random.uniform(1000, 5000, num_rows).round(2),  # بين 1000 و 5000 جنيه
    'Breakdown_Occurrences': np.random.choice([0, 1], num_rows, p=[0.95, 0.05]),  # 5% احتمال حدوث عطل
    'Machine_Efficiency': np.random.uniform(85, 100, num_rows).round(2)  # بين 85% و 100%
}

# إنشاء DataFrame
df = pd.DataFrame(data)
```
